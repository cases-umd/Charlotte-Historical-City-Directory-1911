**<h3>Introduction**

**This is a story about business women. It may be a business with a very specific clientele, but like any business, customers have to be able to find you.**
    
The Charlotte Directory data source in OpenRefine holds 1058 job titles in its database of 15,711 records. Some of these jobs have familiar titles that we still see today (i.e., barber, mason, mill wrkr, student, etc.), and some require historical context to know what the work actually was (i.e., fact hd, turner, warper, wrapper, etc.).
    
One of the jobs listed is "mad" and during a class discussions, it was offered that **"mad"** was short for ***madam***, and that the **madams** ran brothels in Charlotte. Would the public City Directory of Charlotte list the ladies who ruled the vice district? 

**Yes, apparently, it would.**
    
The 29 women who listed "mad" as their job, are a rich source of data about Charlotte in 1911 and about the women themselves, in a profession that was not an easy road to walk (which may be why they operated in alleys). They provide a subset of Charlotte data that includes education, geography and race. 
    
The data also provides a hint at the **restricted areas** that cities considered zones for vice. The police left them alone, as long as they kept to their own "section" of the city.
 




**<h3>Collecting and Manipulating Data**

The 29 records that listed "mad" as a job title were pulled from the larger Charlotte Directory data set. 

This is how a woman would be listed in the Directory.

But "mad" was not in the list of job abbreviations.

I researched prostitution in the early 1900s on the internet. There have been several academic studies, but this site is the most directly and succinctly applicable to this dataset because it specifically references City Directories: 

https://www.pkgraham.com/researching-a-red-light-district-part-1/

It produced this confirmation of directories listing madams in Savannah, GA:

The data were cleaned in OpenRefine to be more consistently formatted, and the data columns were made more granular while keeping the original data for reference.  "Street" was added as a column separate from the address. The empty columns (Title, Married, Wid, Spouse, Company and Biz Directory) were removed to streamline the spreadsheet since they were all empty cells for these records.

There was now a usable set of records to look at. 

**<h3>Visualizing Data**

These women are grouped based on title. They may seem the same, but there is more to their stories.

**Geography:** They mainly lived in clusters on streets,  creating a "red light district" where vice could be found.

They had to stay in their restricted areas, where the police would not arrest them. The police knew they were there, but turned a blind eye as long as they stayed within the unofficial zone boundaries. There is 1 outlier, and that is Kitty West on Boundary Alley. She is the only madam listed in the directory on that street. All of the others are in 4 clusters.

Sanborn Insurance Maps would list the dwellings (D) and other structures as they recorded a city's layout and construction for insurance reports. The abbreviation FB was the abbreviation for "Female Boarding" and was code for a brothel.

This was the largest residential cluster from the dataset. East 1st Street is the only set of addresses for the madams that had the designation "FB" written on it, and then only on one side of the street. The others are simply marked "D" (dwelling).

But that may be because the ones on the south side of the street backed up to local Public School No. 7.

**<h3>1910 U.S. Census - Creating Data**
    
I pulled the 1910 Census records for the Charlotte, Mecklenberg, NC boundaries. I found 14 of the women and increased the data content for these women.

The Census provided a fuller picture (i.e., more data) for each woman.

It was easier to list Dressmaker than to explain your real job to the census taker.

They all said they could read and write.

And they were all listed as "white," but if we look a little closer at the directory, we see some conflicting information:

When cleaning up the first set of data, I removed the "Race" column because they were all listed as "white." The Charlotte directory publishers put an asterisk in front of "black" names. None of the women in the individual listings (our original data source) had an asterisk, so the data listed them as "white." 

The Charlotte Directory has another listing of residents and businesses by street. 
This is the listing of East 1st Street, one of the cluster of addresses in the dataset. All of the "mads" are highlighted. While they are all listed as white, every one of their neighbors is black, illustrating that these restricted districts were located in black neighborhoods. Further research was prompted by the discovery that one of the women in the "mad" dataset had, in this section, an * in front of her name: "807 1/2 * Copeland, Carrie Mad".

Carrie Copeland did not have an asterisk in the residential listing, and several women with the same name appeared in the 1910 census, but one Carrie Copeland was a boarder at William Armstrong's house on East 8th Street. These records may point to the same person because both the directory and census records show the same age range as the other women found with the same job title. 

**<h3>CONCLUSION** 
    
***Yes, it may reasonably be concluded that these women were prostitutes in downtown Charlotte, NC, in 1911, and that they advertised their business in the City Directory. Yes, they lived and worked in restricted areas where prostitution was known to take place, probably with a long list of other vices.***  
    
***If you were looking for them and you were new in town, you could tuck the local City Directory under your arm and start off. If you got lost, you could then actually follow the directions listed on the page.***
